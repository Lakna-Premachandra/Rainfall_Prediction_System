```python
import numpy as np

numerical_cols = train_df.select_dtypes(include=[np.number]).columns.tolist()
for col in ['id', 'rainfall']:
    if col in numerical_cols:
        numerical_cols.remove(col)

skewed_col = 'cloud'

outliers_summary = {}

for col in numerical_cols:
    if col == skewed_col:
        continue
    data = train_df[col]
    z_scores = (data - data.mean()) / data.std()
    outliers = data[(np.abs(z_scores) > 3)]
    outliers_summary[col] = {
        'num_outliers': outliers.count(),
        'outliers_indices': outliers.index.tolist()
    }

Q1 = train_df[skewed_col].quantile(0.25)
Q3 = train_df[skewed_col].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = train_df[(train_df[skewed_col] < lower_bound) | (train_df[skewed_col] > upper_bound)][skewed_col]

outliers_summary[skewed_col] = {
    'num_outliers': outliers_iqr.count(),
    'outliers_indices': outliers_iqr.index.tolist()
}

for col, info in outliers_summary.items():
    print(f"Column: {col}")
    print(f"Number of outliers: {info['num_outliers']}")
    print(f"Outlier indices: {info['outliers_indices']}\n")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind

continuous_vars = ['pressure', 'maxtemp', 'temparature', 'mintemp',
                   'dewpoint', 'humidity', 'cloud', 'sunshine',
                   'winddirection', 'windspeed']

p_values = []

for var in continuous_vars:
    rain_group = train_df[train_df['rainfall'] == 1][var]
    no_rain_group = train_df[train_df['rainfall'] == 0][var]
    stat, p_value = ttest_ind(rain_group, no_rain_group)
    p_values.append(p_value)
    print(f"{var}: p-value = {p_value:.4f} {'(Significant)' if p_value < 0.05 else '(Not significant)'}")

plt.figure(figsize=(12, 6))
colors = ['red' if p < 0.05 else 'blue' for p in p_values]
plt.bar(continuous_vars, p_values, color=colors, alpha=0.7)
plt.axhline(y=0.05, color='red', linestyle='--', linewidth=2, label='Significance Line (0.05)')
plt.title('Feature Significance Test (Red = Significant)', fontsize=14)
plt.xlabel('Features')
plt.ylabel('P-value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

features = train_df.drop(['id', 'rainfall'], axis=1)
labels = train_df['rainfall']

X_test = test_df.drop(['id'], axis=1)

X_train, X_val, y_train, y_val = train_test_split(features, labels, test_size=0.2, random_state=42, stratify=labels)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score, precision_score, recall_score

rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    class_weight='balanced',
    random_state=42,
)
rf_model.fit(X_train, y_train)
```

```python
y_val_pred_rf_proba = rf_model.predict_proba(X_val)[:, 1]
y_val_pred_rf_binary = (y_val_pred_rf_proba >= 0.5).astype(int)

roc_auc_rf = roc_auc_score(y_val, y_val_pred_rf_proba)
print("Validation ROC AUC (Random Forest):", roc_auc_rf)

f1_rf = f1_score(y_val, y_val_pred_rf_binary)
print("Validation F1 Score (Random Forest):", f1_rf)

precision_rf = precision_score(y_val, y_val_pred_rf_binary)
recall_rf = recall_score(y_val, y_val_pred_rf_binary)
print("Precision (Random Forest):", precision_rf)
print("Recall (Random Forest):", recall_rf)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

models_and_params = {
    "Random Forest": {
        "model": RandomForestClassifier(random_state=42),
        "param_grid": {
            'n_estimators': [100, 300, 500],
            'max_depth': [5, 10, None],
            'min_samples_split': [2, 5],
            'min_samples_leaf': [1, 2],
            'class_weight': ['balanced']
        },
        "search": "grid",
        "use_scaled": False
    },
    "MLP": {
        "model": MLPClassifier(random_state=42),
        "param_dist": {
            'hidden_layer_sizes': [(64, 32), (128, 64, 32), (100,), (50, 25)],
            'activation': ['relu', 'tanh'],
            'solver': ['adam', 'lbfgs', 'sgd'],
            'alpha': [0.0001, 0.001, 0.01],
            'max_iter': [300],
            'early_stopping': [True],
            'learning_rate_init': [0.001, 0.01, 0.1]
        },
        "search": "random",
        "n_iter": 20,
        "use_scaled": True
    },
    "XGBoost": {
        "model": XGBClassifier(objective='binary:logistic', eval_metric='auc', use_label_encoder=False, random_state=42),
        "param_grid": {
            'n_estimators': [200, 300],
            'max_depth': [3, 5, 7],
            'learning_rate': [0.01, 0.1],
            'subsample': [0.8, 1],
            'colsample_bytree': [0.8, 1],
            'scale_pos_weight': [1650/540]
        },
        "search": "grid",
        "use_scaled": False
    },
    "LightGBM": {
        "model": LGBMClassifier(random_state=42),
        "param_grid": {
            'n_estimators': [300, 500],
            'max_depth': [6, 8, -1],
            'learning_rate': [0.03, 0.05],
            'num_leaves': [31, 50],
            'subsample': [0.8, 1],
            'colsample_bytree': [0.8, 1],
            'class_weight': ['balanced']
        },
        "search": "grid",
        "use_scaled": False
    },
    "SVM": {
        "model": SVC(probability=True, random_state=42),
        "param_grid": {
            'C': [0.1, 1, 10],
            'gamma': ['scale', 'auto'],
            'kernel': ['rbf', 'poly'],
            'class_weight': ['balanced']
        },
        "search": "grid",
        "use_scaled": True
    }
}

best_models = {}

for name, info in models_and_params.items():
    print(f"Tuning {name}...")
    X_input = X_train_scaled if info["use_scaled"] else X_train
    # GridSearchCV
    if info["search"] == "grid":
        search = GridSearchCV(
            estimator=info["model"],
            param_grid=info["param_grid"],
            scoring='roc_auc',
            cv=3,
            n_jobs=-1,
            verbose=2
        )
    # RandomizedSearch
    else:
        search = RandomizedSearchCV(
            estimator=info["model"],
            param_distributions=info["param_dist"],
            n_iter=info.get("n_iter", 20),
            scoring='roc_auc',
            cv=3,
            n_jobs=-1,
            verbose=2,
            random_state=42
        )

    search.fit(X_input, y_train)
    print(f"Best {name} Params:", search.best_params_)
    print(f"Best {name} ROC AUC (CV):", search.best_score_)

    best_models[name] = {
        "model": search.best_estimator_,
        "use_scaled": info["use_scaled"]
    }
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

n_models = len(best_models)
fig, axes = plt.subplots(1, n_models, figsize=(5 * n_models, 4))

for i, (model_name, model_info) in enumerate(best_models.items()):
    model = model_info['model']
    use_scaled = model_info['use_scaled']

    X_val_input = X_valid_scaled if use_scaled else X_val
    y_val_pred = model.predict(X_val_input)

    cm = confusion_matrix(y_val, y_val_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap='Blues', ax=axes[i] if n_models > 1 else axes, colorbar=False)

    axes[i].set_xlabel("Predicted")
    axes[i].set_ylabel("Actual")
    axes[i].set_title(f"{model_name} - Tuned")

plt.tight_layout()
plt.show()
```

```python
cv_results = {}

for name, info in best_models.items():
    X_cv = X_train_scaled if info["use_scaled"] else X_train
    scores = cross_val_score(info["model"], X_cv, y_train, cv=5, scoring='roc_auc', n_jobs=-1)
    cv_results[name] = scores
    print(f"{name} CV ROC AUC: Mean={np.mean(scores)} Std={np.std(scores)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

model_names = list(cv_results.keys())
mean_scores = [np.mean(cv_results[m]) for m in model_names]
std_scores = [np.std(cv_results[m]) for m in model_names]

plt.figure(figsize=(12, 7))
bars = plt.bar(model_names, mean_scores, yerr=std_scores, capsize=6, color='coral', edgecolor='black')
plt.ylabel('Mean CV ROC AUC', fontsize=14)
plt.title('Comparison of Models Based on 5-Fold Cross-Validated ROC AUC', fontsize=16)
plt.ylim([0.5, 1.05])
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.xticks(rotation=15, fontsize=12)
plt.yticks(fontsize=12)

for bar, mean in zip(bars, mean_scores):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.025, f"{mean:.4f}", ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
best_model_name = max(cv_results, key=lambda k: np.mean(cv_results[k]))
best_model_scores = cv_results[best_model_name]
print(f"Best model: {best_model_name}")
print(f"CV ROC AUC scores: {best_model_scores}")
print(f"Mean CV ROC AUC: {np.mean(best_model_scores)} ± {np.std(best_model_scores)}")
```
